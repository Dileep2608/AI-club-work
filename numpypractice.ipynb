Rearranging Arrays

```python
before = np.array([[1,2,3,4],[5,6,7,8]])
print(before)

after = before.reshape((4,2))
print(after)
```

```python
before = np.array([[1,2,3,4],[5,6,7,8]])
print(before)

after = before.reshape((1,8))
print(after)
```

```python
# Vertically stacking vectors
v1 = np.array([1,2,3,4])
v2 = np.array([5,6,7,8])

np.vstack([v1,v2,v1,v2,v1,v1,v2])
```

```python
# horizantal stacking vectors
v1 = np.array([1,2,3,4])
v2 = np.array([5,6,7,8])

np.hstack([v1,v2,v1,v2])
```

```python
# horizantal stacking vectors
v1 = np.array([[1,2,3,4],[5,6,8,3]])
v2 = np.array([[5,6,7,8],[8,3,4,7]])

np.hstack((v1,v2,v1,v2,v1))
```
